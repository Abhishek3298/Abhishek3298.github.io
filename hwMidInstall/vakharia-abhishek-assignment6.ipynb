```python
#Imporitng the different libarraies for bqplot 
%matplotlib inline

import pandas as pd
import numpy as np
import bqplot
import ipywidgets
import traitlets
```

```python
# Reading the data 
building_data = pd.read_csv('https://uiuc-ischool-dataviz.github.io/is445_AOUAOG_fall2021/week02/data/building_inventory.csv')

#Initially down sampled the data set since there were a lot of data and the system took a lot of time to respond and get the
#required output
#nsamples = 5000
#downSampleMask = np.random.choice(range(len(building_data)-1), nsamples, replace=False)
#DS = building_data.loc[downSampleMask]
```

```python
#Grouping the data by Countr and Agency Name while getting the mean of the square footage
dataDS = building_data.groupby(['County', 'Agency Name'])['Square Footage'].mean()
#Using unstack to rearrange the data in a matrix format with rows and columns
dataDS = dataDS.unstack(level=0).round(2)
```

# Heat Map Creation and Bar Plot 

-I initially transformed the data using the groupby() function to structure the data for the heat map

-Then I plotted a heat map and implemeted the interactions 

-We has to assign the Agency Name as the row argument and County as the column argument

-I used the red color schema for the heat map and also used the log function to create a gradient according to the weight 

-Then created a simple barplot for the mean square footage by the year they were acquired in

```python
# Declaring the lable widgets for the heat map which prints the text 
mean_label = ipywidgets.Textarea()
mean_label.value = "Mean Sq. footage of the selected criteria: No data selected!"
mean_label.layout.width = '100%'

#the color for the heat map - red schema which was referred from the class notbook 
color_sc = bqplot.ColorScale(scheme= 'Reds') 
#the differnt scales for the heatmap
x_scale = bqplot.OrdinalScale()
y_scale = bqplot.OrdinalScale()
# Axes for heat map
ax_x = bqplot.Axis(scale = x_scale, tick_style= {'font-size': '8px'}, tick_rotate = 75)
ax_y = bqplot.Axis(scale = y_scale, tick_style= {'font-size': '7px'}, orientation = 'vertical')
#displaying the colour gradien on the right side of the heat map
ax_color = bqplot.ColorAxis(scale = color_sc, orientation = 'vertical', side = 'right')


# Heat map object
# using the log function to create a gradient of colors
hm = bqplot.GridHeatMap(color = np.log10(dataDS.to_numpy()), row = dataDS.index.tolist(), column = dataDS.columns.tolist(),
                                scales = {'color': color_sc, 'row': y_scale, 'column': x_scale}, interactions = {'click': 'select'},
                                anchor_style = {'fill': 'neon'}, selected_style={'stroke': 'neon', 'stroke-width': 3},
                                unselected_style={'opacity':0.8})

#Setting the layout width to 50% since we want to display the barplot side by side
figure = bqplot.Figure(marks = [hm], axes = [ax_color, ax_x, ax_y])
figure.layout.width = '50%'
```

```python
# The different scales for the bar plot - x & y
xb_scale = bqplot.OrdinalScale()
yb_scale = bqplot.LinearScale()

# Axes for bar plot
ax_xb = bqplot.Axis(scale = xb_scale, tick_rotate = 90, label = 'Year')
ax_yb = bqplot.Axis(scale = yb_scale, orientation = 'vertical', label = 'Total Square Footage')

# Bar plot object
bp = bqplot.Bars(x = [], y = [], scales={'x': xb_scale, 'y': yb_scale})

#setting the layout width to 50% since we want to display with the heat map side by side
figureb = bqplot.Figure(marks = [bp], axes=[ax_xb, ax_yb])
figureb.layout.width = '50%'
```

- After working on the heat map and the bar plot, we had to link the two 
- When a cell is clicked on the heat map , it creates a corresponding bar plot 
- The Mean of the Square Footage is also displayed on the top left corner along with the number of cells selected 

```python
# Creating the interactivity on the heat map which generates a bar plot for the corresponding cell
def on_select(self, change, dataDS, building_data):
        govt_dept = []
        county = []
        if len(self.selected) == 0:
            mean_label.value = 'No Data Selected!'
            return
        m_label = 'Mean Sq. Footage of '+str(len(self.selected))+' selected criteria: \n'    
            
        for i in self.selected:
            m_label += f'{dataDS.index[i[0]]}, {dataDS.columns[i[1]]}: {dataDS.iloc[i[0], i[1]]}\n'
            govt_dept.append(dataDS.index[i[0]])
            county.append(dataDS.columns[i[1]])
            
        data_sub = building_data.loc[building_data['Agency Name'].isin(govt_dept) & building_data['County'].isin(county), :]
        data_sub = data_sub.groupby('Year Acquired')['Square Footage'].sum().reset_index()
        
        if data_sub.empty:
            bp.x = []
            bp.y = []
        
        mean_label.value = m_label    
        bp.x = data_sub['Year Acquired'].to_numpy()
        bp.y = data_sub['Square Footage'].to_numpy()
                
# using the lambda function to link the interaction 
hm.on_element_click(lambda _,x: on_select(_, x, dataDS, building_data))
```

```python
# This is used to display the text box along with the plots 
db_interactive = ipywidgets.VBox([mean_label, ipywidgets.HBox([figure, figureb])])
db_interactive
```
